DESAFIO 3

Você recebeu um dicionário onde:

*   a chave representa o nome da participante;
*   o valor representa o status de conclusão do desafio (True ou False).

Tarefa: Crie um programa que:

*   Conte quantas participantes concluíram o desafio;
*   Exiba apenas o nome das participantes que não concluíram.

```python
#Execute seu código aqui
participantes = {
    "Ana" : True,
    "Natália": False,
    "Anna Luiza": False,
    "Soniany": True,
    "Julie": True,
    "Pandora": True
}

qtd_status_concluidos = 0

for nome, status in participantes.items():
  if status == True:
    qtd_status_concluidos += 1

print("Total de concluídos: ", qtd_status_concluidos)

print("Participantes que não concluíram:")
for nome, status in participantes.items():
  if status == False:
    print(nome)
```

DESAFIO 4


Você recebeu uma lista com registros de acesso à plataforma (horas em formato inteiro).

Tarefa: Crie um programa que:

Conte quantos acessos ocorreram em cada período:

Manhã (6–11)

Tarde (12–17)

Noite (18–23)

Exiba o total por período.

```python
#Execute seu código aqui
acessos = [9, 8, 5, 12, 22, 20, 18, 14, 16, 7, 17]

manha = 0
tarde = 0
noite = 0

for hora in acessos:
  if 6 <= hora <= 11:
    manha+=1
  elif 12 <= hora <= 17:
    tarde+=1
  elif 18 <= hora <= 23:
    noite+=1

print("Manhã:", manha)
print("Tarde:", tarde)
print("Noite:", noite)
```
